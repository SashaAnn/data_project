```python
def clean_sh(sleep_health) -> pd.DataFrame:
    # Clean column names
    sleep_health.columns = sleep_health.columns.str.title()
    sleep_health.columns = sleep_health.columns.str.strip().str.replace('_', ' ')

    # List of columns to drop
    cols_to_drop = ['Person ID', 'Age', 'Occupation', 'Stress Level', 'BMI Category', 'Blood Pressure', 'Heart Rate', 'Daily Steps', 'Sleep Disorder']

    # Drop columns only if they exist in the DataFrame
    sleep_health = sleep_health.drop(columns=[col for col in cols_to_drop if col in sleep_health.columns])

    # Drop rows with missing values
    sleep_health = sleep_health.dropna()

    return sleep_health
```

```python
# New column
def categorize_activity(minutes):
    if minutes < 30:
        return 'low'
    elif 30 <= minutes <= 60:
        return 'moderate'
    else:
        return 'high'
    
sleep_health['Physical Acitivity Level'] = sleep_health['Physical Activity Level'].apply(categorize_activity)
```

```python
# calling function to clean data
sleep_health_cleaned = clean_sh(sleep_health)
sleep_health_cleaned.head(5)
```

```python
# Sleep Duration Stats
duration_mean = sleep_health['Sleep Duration'].mean()
duration_median = sleep_health['Sleep Duration'].median()
duration_min = sleep_health['Sleep Duration'].min()
duration_max = sleep_health['Sleep Duration'].max()

print(f'The mean number of Sleep Duration is {duration_mean}.\n The median number of Sleep Duration is {duration_median}.\n The minimum number of Sleep Duration is {duration_min}.\n The maximum number of Sleep Duration is {duration_max}.')
```

```python
# Quality Of Sleep Stats
quality_mean = sleep_health['Quality Of Sleep'].mean()
quality_median = sleep_health['Quality Of Sleep'].median()
quality_min = sleep_health['Quality Of Sleep'].min()
quality_max = sleep_health['Quality Of Sleep'].max()

print(f'The mean number of Quality Of Sleep is {quality_mean}.\n The median number of Quality Of Sleep is {quality_median}.\n The minimum number of Quality of Sleep is {quality_min}.\n The maximum number of Quality Of Sleep is {quality_max}.')
```

```python
# Physical Activity Level Stats
activity_mean = sleep_health['Physical Activity Level'].mean()
activity_median = sleep_health['Physical Activity Level'].median()
activity_min = sleep_health['Physical Activity Level'].min()
activity_max = sleep_health['Physical Activity Level'].max()

print(f'The mean number of Physical Activity Level is {activity_mean}.\n The median number of Physical Activity Level is {activity_median}.\n The minimum number of Physical Activity Level is {activity_min}.\n The maximum number of Physical Activity Level is {activity_max}.')
```

```python
# Group the data
grouped_data = sleep_health.groupby(['Gender', 'Sleep Duration'])['Quality Of Sleep'].mean().reset_index()

# Pivot the data for easier plotting
pivot_data = grouped_data.pivot(index='Sleep Duration', columns='Gender', values='Quality Of Sleep')

# Create the bar plot
plt.figure(figsize=(12, 6))

x = pivot_data.index
bar_width = 0.4
x_indices = range(len(x))

# Plot bars for Male and Female
plt.bar([i - bar_width / 2 for i in x_indices], pivot_data['Male'], width=bar_width, label='Male', color='blue')
plt.bar([i + bar_width / 2 for i in x_indices], pivot_data['Female'], width=bar_width, label='Female', color='orange')

# Customize the plot
plt.xticks(ticks=x_indices, labels=x, rotation=90)
plt.title('Average Quality Of Sleep by Sleep Duration and Gender', fontsize=16)
plt.xlabel('Sleep Duration', fontsize=14)
plt.ylabel('Average Quality Of Sleep', fontsize=14)
plt.legend(title='Gender')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Display the plot
plt.show()
```

Analysis and Insights:
1. Both males and females generally experience higher sleep quality with longer sleep durations, peaking around 8 to 8.5 hours.
2. For sleep durations between 5.8 and 6.4 hours, females tend to have a higher average quality of sleep compared to males.
3. Males generally report higher sleep quality than females from 7 to 7.5 hours.
4. For sleep durations of 7.5 hours and above, females report higher sleep quality than males.
5. Females reach the highest average quality of sleep at sleep durations of 8 to 8.5 hours.
6. Males reach the highest average quality of sleep at sleep durations of 7 to 7.5 hours.


Conclusion:

Slightly different sleep durations might be optimal for different genders in terms of sleep quality, as evidenced by the highest average quality of sleep achieved for females and males.

Females generally report higher sleep quality than males for most sleep durations.

Both genders show an increase in sleep quality with longer sleep durations.

These conclusions suggest that individualized approaches to sleep health might be beneficial.
